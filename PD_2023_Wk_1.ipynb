```python
#Importing pandas for data cleaning and manipulation
import pandas as pd
```

```python
#Importing the dataset
pd1 = pd.read_csv(r'/Users/KellyLam/Desktop/Data_Analysis/Data Cleaning/PD 2023 Wk 1 Input.csv')
```

```python
pd1.head()
```

```python
#Extracting letters from the Transaction Code and renaming column to Bank
Bank = pd1['Transaction Code'].str.extract(r'^([A-Za-z]+)')
pd1['Transaction Code'] = Bank
pd1 = pd1.rename(columns={'Transaction Code': 'Bank'})
```

```python
pd1.head()
```

```python
#Renaming values in Online or In-Person column from 1 and 2 to Online and In-Person
rename = {1: 'Online', 2: 'In-Person'}
pd1['Online or In-Person'] = pd1['Online or In-Person'].replace(rename)
```

```python
pd1.head()
```

```python
#Changing the date to be the day of the week
pd1['Transaction Date'] = pd.to_datetime(pd1['Transaction Date']).dt.day_name()
```

```python
pd1.head()
```

```python
#Creating output of Total values of Transactions by each bank
bank_sum = pd1.groupby('Bank')['Value'].sum()
df_bank_sum = pd.DataFrame(bank_sum).reset_index()
```

```python
df_bank_sum.head()
```

```python
#Creating output of Total Values by Bank, Day of the Week, and Type of Transaction
bwt_sum = pd1.groupby(['Bank', 'Transaction Date', 'Online or In-Person'])['Value'].sum()
df_bwt_sum = pd.DataFrame(bwt_sum).reset_index()
```

```python
df_bwt_sum.head()
```

```python
#Creating output of Total Values by Bank and Customer Code
bc_sum = pd1.groupby(['Bank', 'Customer Code'])['Value'].sum()
df_bc_sum = pd.DataFrame(bc_sum).reset_index()
```

```python
df_bc_sum.head()
```

```python
#Exporting each dataframe to CSV
folder_path = '/Users/KellyLam/Desktop/Data_Analysis/Data_Cleaning/'
file_name1 = 'bank_totals_by_bank.csv'
file_path1 = folder_path + file_name1

df_bank_sum.to_csv(file_path1, index=False)
```

```python
folder_path = '/Users/KellyLam/Desktop/Data_Analysis/Data_Cleaning/'
file_name2 = 'bank_totals_transaction_type.csv'
file_path2 = folder_path + file_name2

df_bwt_sum.to_csv(file_path2, index=False)
```
